NUMPY INDEXING (accessing data)

indexing : selecting a specific value from an array

ex : picking one student from a class list


```python
# indexing in 1D array (index)

import numpy as np
arr1=np.array([1,2,3,4])
print(arr1[0])
print(arr1[2])
```

```python
# indexing in 2D array (row,column)
arr2=np.array([[1,2],
               [3,4]])
print(arr2[0,1])
```

```python
# indexing in 3D array (layer,row,column)
arr3=np.array([[[1,2,3],
                [4,5,6]],
                [
                [6,7,8],
                [7,8,9]
                ]])
print(arr3[0,1,2])
```
